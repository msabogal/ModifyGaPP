$$Estimation from D(z) data$$

```python
# load the {z,ln(D)} data from Hu Zhan (2009) http://dx.doi.org/10.1088/0004-637X/690/1/923
(Z0,SD) = loadtxt("./LSST_simulated.txt", unpack='True')
```

```python
parametros = {'Omega_m' : 0.3,'H0' : 70, 'alpha' : 1.0, 'beta' : 0.69,  'ini_sigma':10.0, 'ini_lpar':10.0,
              'xmin': 0.0, 'xmax': 4.1, 'nstar' : 300, 'N_rec': 100, 'N_simulated_by_rec': 1000,
              'cov_fun': covariance.SquaredExponential,'fgrad': 'True', 'fiducial': 'LCDM'}

case1= Gp_Analisys_D(Z0,SD,parametros) #Asignacion de condicione siniciales y parametros 

case1.N_models()                       #Genera N conjutos de datos simulados 
case1.N_reconstructions()              #Reconstruccion de los N conjutos de datos simulados   
```

```python
case1.statistics(print_stad='True') #Se realiza el tratamiento estadistico con los valores obtinodos de las rec_ 
```

```python
case1.plots(x_lims=[0,1.0],y_lims=[65,125])
```
